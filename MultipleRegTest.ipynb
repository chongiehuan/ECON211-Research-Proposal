```python
## Import dataset of variables chosen from multivariate analysis
## py = EU-ETS Carbon Price, px1 = Brent Crude Prices, px2 = DAX Index Prices
## Data sorted based on availability from 01/01/2010 to 31/12/2018
## Link to data: https://raw.githubusercontent.com/chongiehuan/ECON211-Research-Proposal/main/CarbonPrice.csv?token=ARRX544PLSCALWCIYEXBNBTAIXLXI
import csv
f = open('CarbonPrice.csv')
csv_f = csv.reader(f)

for row in csv_f:
  print (row)
```

```python
df = pandas.read_csv("CarbonPrice.csv")
```

```python
X = df[['px1', 'px2']]
y = df['py']
```

```python
from sklearn import linear_model
```

```python
#Creating the multiple regression and producing fit score
regr = linear_model.LinearRegression()
regr.fit(X, y)
regr.score(X,y)
```

```python
## Fit score is extremely weak, this is partly because of a lack of sufficient explanatory variables.
## Additionally, this is largely the problem with statistical regression models: they have poor predictive power.
```

```python
## Example of Model Prediction Outcome with Given Inputs
predictedCarbonPrice = regr.predict([[100, 7000]])
print (predictedCarbonPrice)
```

```python
regr = linear_model.LinearRegression()
regr.fit(X, y)

print(regr.coef_)
```

```python
## As hypothesised, the coefficient for energy futures is negatively related to the Carbon Price.
```
